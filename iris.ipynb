```python
import keras
from tensorflow.keras import Sequential
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import LabelBinarizer
from sklearn.datasets import load_iris
```

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# loading the dataset
data = load_iris()
```

```python
data
```

```python
#creating the data frames
data = pd.DataFrame(data= np.c_[data['data'], data['target']],
                     columns= data['feature_names'] + ['target'])
```

```python
data
```

```python
sns.pairplot(data=data[data.columns[0:6]], hue='target')
plt.show()
```

HeatMap

```python
plt.figure(figsize=(8,5)) 
sns.heatmap(data.corr(),annot=True,cmap='cubehelix_r') 
plt.show()
```

```python
data 
```

```python
X=data[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
target=data[['target']]
y_true=target.copy()
```

```python
X
```

```python
target
```

Train the model

```python
X = StandardScaler().fit_transform(X)
target = LabelBinarizer().fit_transform(target)
```

Dividing the train and testing set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, target, test_size = 0.4, random_state = 101)
```

Model

We are defining the epoch 

1 epoch is is for one round of forword to backword 

```python
model = Sequential()
model.add(Dense( 12, input_dim=4, activation = 'relu'))
model.add(Dense( units = 15, activation= 'relu'))
model.add(Dense( units = 8, activation= 'relu'))
model.add(Dense( units = 10, activation= 'relu'))
model.add(Dense( units = 3, activation= 'softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
history = model.fit(x_train, y_train, epochs = 170, validation_data = (x_test, y_test))
```

Preding the testing result

```python
y_test
```

```python
y_test=np.argmax(y_test, axis=1)
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test.tolist(), y_test.tolist())

import seaborn as sns
import matplotlib.pyplot as plt     

ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['setosa', 'versicolor', 'virginica']); ax.yaxis.set_ticklabels(['setosa', 'versicolor', 'virginica']);
```

```python
from sklearn.metrics import precision_recall_fscore_support
```

```python
precision_recall_fscore_support(y_test.tolist(), y_test.tolist(), average='macro')
```

Ploting accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Accuracy")
plt.legend(['train', 'test'])
plt.show()
```

The variation shows how the data values are tested 

the orange variation shows how much values are wrongly predicted

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss')
plt.legend(['Train','Test'])
plt.show()
```
